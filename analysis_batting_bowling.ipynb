#  Data Analysis on Player Statistics in Cricket

The dataset consists of Player Statistics in ODI, T20 and Test matches. We would like to group players into "similar" groups by Kmeans clustering algorithms. The cricket players are classified into:  
            1. Best batsmen 
            2. Good (but not best) batsmen
            3. Best bowlers
            4. Good (but not best) bowlers
            5. Best other contributors

Dataset for T20 batting and fielding

> d <- read.csv("twenty20careerbattingandfielding.csv")
> library(tidyr)
> d <- drop_na(d)
> x <- kmeans(d$Strike_Rate,3)
> x$cluster <- as.factor(x$cluster)
> o <- order(x$cluster)
> cluster_elements <- data.frame(d$Name[o], x$cluster[o])
> write.csv(cluster_elements, "Cluster_no.csv")
> ggplot(d, aes(Average, Strike_Rate, col = x$cluster)) + geom_point()

Applying Kmeans clustering on the given data, we obtain 3 clusters. Players in the Cluster1 are classified as the Best batsman and players in the Cluster2 as Good(but not the best) batsman.
